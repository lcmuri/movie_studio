```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./cleaned_data.csv')
data
```

```python
# Convert lists to NumPy arrays
grossings = np.array([float(value.replace("$", "").replace(",", "")) for value in data['worldwide_gross']])
budget = np.array([float(value.replace("$", "").replace(",", "")) for value in data['production_budget']])

roi = grossings/budget

roi = pd.Series(roi)
roi
```

```python
data['roi'] = roi
data
```

```python
data.to_csv('./addedroi.csv')
```
