```python
import sqlite3
import pandas as pd
```

```python
##paths to relevant datasets
imdb = './Group_Project/DataUnzipped/im.db'
bom = './Group_Project/DataUnzipped/bom.movie_gross.csv'
tmdb = './Group_Project/DataUnzipped/tmdb.movies.csv'
tn = './Group_project/DataUnzipped/tn.movie_budgets.csv'
```

```python
##Database data
con = sqlite3.connect(imdb)

query = '''
SELECT * FROM movie_basics
'''

db = pd.read_sql_query(query, con)
db
```

```python
names = db.iloc[:,1]
names
```

```python
#bom_data
bomojo = pd.read_csv(bom)
bomojo.head()
```

```python
titles = bom.iloc[:,0]
```

### Creating a mask to check for matching names

```python
# Create a mask for matching names
mask = titles.isin(names)

positive_matches = titles[mask]
# Get corresponding positive values
#positive_matches = values_list[mask & (values_list > 0)].tolist()

# print(positive_matches)
len(positive_matches)
```

```python
obs = positive_matches.to_list()
```

```python
tm = pd.read_csv('./Group_Project/DataUnzipped/tmdb.movies.csv')
tls = tm.iloc[:,7]
tls
```

```python
budget = pd.read_csv('./Group_project/DataUnzipped/tn.movie_budgets.csv')
movies = budget.iloc[:,2]
budget.head()
```

```python

vals = titles.isin(tls)
matches = titles[vals]
matches
# len(matches)
```

```python
check = matches.isin(movies)
new = matches[check]
len(new)
```

```python
v = new.isin(tls)
v.isna().count()
```

```python
filtered_budget = budget[budget['movie'].isin(new)]
filtered_budget = filtered_budget.rename(columns={'movie':'name'})
# filtered_budget
```

```python
filtered_tm = tm[tm['title'].isin(new)]
filtered_tm = filtered_tm.rename(columns={'title':'name'})
# filtered_tm
```

```python
filtered_bomojo = bomojo[bomojo['title'].isin(new)]
filtered_bomojo = filtered_bomojo.rename(columns={'title':'name'})
# filtered_bomojo
```

```python
filtered_db = db[db['primary_title'].isin(new)]
filtered_db = filtered_db.rename(columns={'primary_title':'name'})
# filtered_db
```

```python
merged_df = pd.merge(filtered_bomojo, filtered_db, on='name', how='inner')
merged_df
```

```python
doublemerged_df = pd.merge(merged_df, filtered_tm, on='name', how='inner')
doublemerged_df.info()
```

```python
fullmerge_df = pd.merge(doublemerged_df, filtered_budget, on='name', how='inner')
fullmerge_df.info()
```

```python
fullmerge_df.to_csv('./Group_Project/mergedData.csv')
```
