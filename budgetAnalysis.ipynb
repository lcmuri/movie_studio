```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./addedroi.csv')
```

```python
data
```

```python
data['studio']
```

```python
## Group by month
studio_roi = data.groupby('studio')['roi'].mean().reset_index()
```

```python
studio_roi
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.scatter(studio_roi['studio'], studio_roi['roi'], marker='o')
plt.title('Average ROI by Budget')
plt.xlabel('roi')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
